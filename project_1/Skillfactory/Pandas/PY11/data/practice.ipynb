```python
def get_experience(arg):
    month_key_words = ['месяц', 'месяцев', 'месяца']
    year_key_words = ['год', 'лет', 'года']
    args_splited = arg.split(' ')
    month = 0
    year = 0
    for i in range(len(args_splited)):
        if args_splited[i] in month_key_words:
            month = args_splited[i-1]
        if args_splited[i] in year_key_words:
            year = args_splited[i-1]
    return int(year)*12 + int(month)
print(get_experience('опыт работы 8 лет и 3 месяца'))
```

```python
citibike_df['starttime'] = pd.to_datetime(citibike_df['starttime'])
citibike_df['stoptime'] = pd.to_datetime(citibike_df['stoptime'])
citibike_df['trip duration'] = citibike_df['stoptime'].dt.minute - citibike_df['starttime'].dt.minute
display(citibike_df)
```

```python
weekday_trip = citibike_df['starttime'].dt.dayofweek
def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        return 1
    else:
        return 0  
citibike_df['Weekend'] = weekday_trip.apply(get_weekend)
result = citibike_df[citibike_df['Weekend'] == 1].shape
display(result)
display(citibike_df)
```

```python
time_hour = citibike_df['starttime'].dt.hour
def time_of_day_func(time):
    if 0 <= time <= 6: return 'night'
    if 6 < time <= 12: return 'morning'
    if 12 < time <= 18: return 'day'
    if 18 < time <= 23: return 'evening'
citibike_df['time_of_day'] = time_hour.apply(time_of_day_func)
day = citibike_df[citibike_df['time_of_day'] == 'day'].shape[0]
night = citibike_df[citibike_df['time_of_day'] == 'night'].shape[0]
display(round(day / night))
```
