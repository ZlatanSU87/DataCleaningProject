```python
films_df = pd.read_csv('ratings_movies.csv')
display(films_df)
```

```python
# Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
films_df['year_release'] = films_df['title'].apply(get_year_release)
films_df['year_release'] = films_df['year_release'].astype('Int64')
year = films_df['year_release'] == 1999
films_df[year].groupby('title')['rating'].mean().sort_values(ascending=True)
```

```python
genres_list = films_df['year_release'] == 2010
films_df[genres_list].groupby('genres')['rating'].mean().sort_values(ascending=True)
```
