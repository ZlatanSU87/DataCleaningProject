```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
churn_data = pd.df = pd.read_csv("churn.csv")
churn_data = churn_data.drop('RowNumber', axis=1)
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"
churn_data['CreditScoreCat'] = churn_data['CreditScore'].apply(get_credit_score_cat)
churn_data
pivot_data = churn_data.pivot_table(
    values='Exited',
    columns='Tenure',
    index='CreditScoreCat',
    aggfunc='mean'
)
pivot_data
fig = sns.heatmap(
    data=pivot_data,
    cmap='YlOrBr',
)
fig.set_title('Тепловая карта оттока клиентов', fontsize=16);
```

#### Судя по карте наибольший отток клиентов наблюдается в категории 'very Poor', на втором месте - 'Excellent'. При этом, бросается в глаза большой отток до года использования услуг банка, 3-4 года (в случае 'very Poor'), от 8 и до 10 лет. Меньший отток клиентов наблюдается в диапазоне 'Poor' - 'Fair' вне значительной зависимости от времени пользования услугами. Банку следует поработать над мотивацией новых клиентов, а также пользующимися услугами свыше 8 лет.
