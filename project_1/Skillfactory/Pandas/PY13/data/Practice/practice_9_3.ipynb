```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
churn_data = pd.df = pd.read_csv("churn.csv")
churn_data = churn_data.drop('RowNumber', axis=1)
churn_data


fig = px.box(
    data_frame=churn_data,
    x='Balance',
    y='Exited', 
    color='Exited',
    orientation='h',
    width=800,
    height=500,
    title='Баланс клиентов в зависимости от оттока'
)

fig.show()
```

#### На указанном графике видно, что среди убывших клиентов большинство имеют на балансе сумму в размере от 37 до 137 тыс долларов (медиана - 109 тыс долларов). Диапозон оставшихся клиентов больше и в большинстве представлен от минимального значения - 0, до 126 тыс долларов на балансе (медиана - 92 тыс долларов). Очевидная причина ухода клиентов не прослеживается. Нужно обратить внимание на другие показатели. Возможно, ушедших клиентов не устраивают условия содержания банковского счета при балансе свыше 100 тыс долларов. К примеру, высокая комиссия или ограничение при операциях с большими суммами.
