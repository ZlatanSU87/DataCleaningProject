```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
churn_data = pd.df = pd.read_csv("churn.csv")
churn_data = churn_data.drop('RowNumber', axis=1)
churn_data


fig = px.box(
    data_frame=churn_data,
    x='Age',
    y='Exited', 
    color='Exited',
    orientation='h',
    width=800,
    height=500,
    title='Лояльность клиентов в зависимости от возраста'
)

fig.show()
```

#### На указанном графике больше всего выбросов наблюдается в группе оставшихся клиентов (непрерывно с возраста от 57 до 85 лет). Также прослеживается тенденция, согласно которой большая часть оставшихся клиентов моложе, чем группа ушедших (медианные значения 36 и 45 соответственно). Вместе с тем, разница в возрасте кардинально не отличается. Банку следует обратить внимание на возрастную категорию клиентов от 40 лет.
