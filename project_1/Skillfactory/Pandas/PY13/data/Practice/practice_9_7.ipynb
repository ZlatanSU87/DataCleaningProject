```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
churn_data = pd.df = pd.read_csv("churn.csv")
churn_data = churn_data.drop('RowNumber', axis=1)
mask = churn_data['Exited'] == 1
churn_data_ex = churn_data[mask].groupby('NumOfProducts', as_index=False).count()
churn_data_ex
fig = sns.barplot(
    data=churn_data_ex,
    y='Exited',
    hue='NumOfProducts',
    orient='v',   
)
fig.set_title('Отток клиентов в зависимости от количества приобретенных услуг банка');
```

Банку следует обратить внимание на условия приобретения большего числа услуг одним клиентом, явно есть факторы, отталкивающие от увеличения количества услуг.
