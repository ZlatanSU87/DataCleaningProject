```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
churn_data = pd.df = pd.read_csv("churn.csv")
churn_data = churn_data.drop('RowNumber', axis=1)
mask = churn_data['Exited'] == 1
churn_data_ex = churn_data[mask].groupby('IsActiveMember', as_index=False).count()
churn_data_ex['IsActiveMember'] = churn_data_ex['IsActiveMember'].apply(lambda x: 'Неактивный' if x == 0 else 'Активный')
fig = sns.barplot(
    data=churn_data_ex,
    y='Exited',
    hue='IsActiveMember',
    orient='v',   
)
fig.set_title('Отток клиентов в зависимости от статуса активного клиента');
```

#### Данный график свидетельствует о том, что доля ушедших клиентов, не имеющих активного статуса, почти в 2 раза превышает долю ушедших клиентов, имеющих активный статус. Банку следует акцентировать внимание на дополнительную мотивацию клиентов в получении активного статуса.
